```python
from datetime import datetime as dt
```

```python
def _fib(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else: 
        return _fib(n-1) + _fib(n-2)

def fib(N):
    now = dt.now()
    print(f"starting at {now}")
    fdict = {}
    for n in range(N):
        # print()
        now1 = dt.now()
        # print(f"{n}th Fibonacci # ==>{_fib(n)}      {str(now1 - now)}")
        fdict[_fib(n)] = str(now1 - now)
    return fdict
```

```python
import pandas as pd

fd = fib(35)
# print(fd)
fnums,tt = [],[]
for f,t in fd.items():
    fnums.append(f)
    tt.append(t)
# print(fnums)
# print(tt)
dfib = pd.DataFrame.from_dict({'fib':fnums,'ttaken':tt})
# print(dfib)

import matplotlib.pyplot as mp
# mp.plot(dfib)

dfib['milis'] = [int(x[9:]) for x in dfib['ttaken']]
pdf = pd.DataFrame()
pdf['fib'] = dfib['fib']
pdf['milis'] = dfib['milis']
mp.xlabel('Fibonacci Num')
mp.ylabel('Milliseconds Taken')

# mp.plot(dfib['fib'],dfib['milis'])
mp.xticks(dfib['fib'])
mp.plot(pdf)
mp.show()
```

```python
def genFib(N):
    result = [0,1]
    for i in range(2,N):
        # print(f"result[i-2] = {result[i-2]}")
        # print(f"result[i-1] = {result[i-1]}")
        result.append(result[i-2] + result[i-1])
    return result

print(genFib(50))
```

```python
from numpy import roots, arange
r1,r2 = roots([1,-1,-1])
print(r1)
print(r2)
C = 1.0 / (r1-r2)

def fib2s(N):
    n = arange(N,dtype=float)
    print(n)
    return C*(r1**n - r2**n)

x = fib2s(5)
print(x)
```
