```python
def _fib(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else: 
        return _fib(n-1) + _fib(n-2)

def fib(N):
    now = dt.now()
    print(f"starting at {now}")
    fdict = {}
    for n in range(N):
        # print()
        now1 = dt.now()
        # print(f"{n}th Fibonacci # ==>{_fib(n)}      {str(now1 - now)}")
        fdict[_fib(n)] = str(now1 - now)
    return fdict
```

```python
import pandas as pd

fd = fib(35)
# print(fd)
fnums,tt = [],[]
for f,t in fd.items():
    fnums.append(f)
    tt.append(t)
# print(fnums)
# print(tt)
dfib = pd.DataFrame.from_dict({'fib':fnums,'ttaken':tt})
# print(dfib)

import matplotlib.pyplot as mp
# mp.plot(dfib)

dfib['milis'] = [int(x[9:]) for x in dfib['ttaken']]
pdf = pd.DataFrame()
pdf['fib'] = dfib['fib']
pdf['milis'] = dfib['milis']
mp.xlabel('Fibonacci Num')
mp.ylabel('Milliseconds Taken')

# mp.plot(dfib['fib'],dfib['milis'])
mp.xticks(dfib['fib'])
mp.plot(pdf)
mp.show()
```
