```python
x = pd.DataFrame(df.YearsExperience)

y = pd.DataFrame(df.Salary)
print(type(x))
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42) # splits the datasets into training and testing data

```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)

plt.figure(figsize=(12,7))
plt.title('Prediction plot for training dataset')
plt.xlabel('Years of Experince')
plt.ylabel('Salary')
plt.scatter(X_train,y_train)
plt.plot(X_test,y_pred)
plt.show()
```

```python


coeff = lr.coef_
intercept = lr.intercept_

print(f"coefficient => {coeff}")
print(f"intercept => {intercept}")
```
